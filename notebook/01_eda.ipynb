```python
import pandas as pd

url = "https://raw.githubusercontent.com/DunjaPawlovic/product-category-classification/main/data/products.csv"

df = pd.read_csv(url)

df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df_clean = df.dropna(subset=["Category Label", "Product Title"])
```

```python
df_clean.isnull().sum()
```

```python
len(df), len(df_clean)
```

```python
X = df_clean["Product Title"]
y = df_clean["Category Label"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
X_vec = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```
